# Punto 1:

- `[15 pts]` Hacer Web Scraping de 10 animales en Wikipedia (en búcle)
- `[10 pts]` Obtener el **encabezado** de cada animal
- `[15 pts]` Obtener todos los **textos** que están en las etiquetas de negrilla.

```python
import bs4 as bs 
import urllib.request
```

```python
enlaces = ["https://es.wikipedia.org/wiki/Carcharocles_megalodon", "https://es.wikipedia.org/wiki/Triceratops", 
          "https://es.wikipedia.org/wiki/Brontosaurus", "https://es.wikipedia.org/wiki/Velociraptor",
          "https://es.wikipedia.org/wiki/Tyrannosaurus_rex", "https://es.wikipedia.org/wiki/Allosaurus",
          "https://es.wikipedia.org/wiki/Stegosaurus","https://es.wikipedia.org/wiki/Brachiosaurus_altithorax",
          "https://es.wikipedia.org/wiki/Pterodactylus","https://es.wikipedia.org/wiki/Quetzalcoatlus",
          "https://es.wikipedia.org/wiki/Soricidae"]
```

```python
# Buscar encabezados y listarlos 
i=1
for link in enlaces:
    request = urllib.request.Request(link, headers={'User-Agent': 'Mozilla/5.0'})
    webpage = urllib.request.urlopen(request)
    source = webpage.read()
    webpage.close()
    soup = bs.BeautifulSoup(source, 'html.parser')
    encabezado = soup.find("body").find_all("h1")[0].text
    print(f'{i}. {encabezado}')
    i+=1
```

```python

for link in enlaces:
    request = urllib.request.Request(link, headers={'User-Agent': 'Mozilla/5.0'})
    webpage = urllib.request.urlopen(request)
    source = webpage.read()
    webpage.close()
    soup = bs.BeautifulSoup(source, 'html.parser')
    negritas = soup.find("body").find_all("b")
    listadoNegritas =[]    
    for val in negritas:
        listadoNegritas.append(val.text)
    
    print(f'-------------------------------------------------------')
    print(f'Página Web: {link}')
    print(f'-------------------------------------------------------')
    print(f'Palabras en negrilla:')
    print(f'-------------------------------------------------------')
        
    i=1    
    for palabra in listadoNegritas:
        print(f'{i}. {palabra}')
        i+=1
```

# Punto 2:
- `[10 pts]` Usando regex, reemplazar todas las tíldes del punto anterior por un asterisco (¡Ojo, los espacios se quedan!)


```python
import re
```

```python

for link in enlaces:
    request = urllib.request.Request(link, headers={'User-Agent': 'Mozilla/5.0'})
    webpage = urllib.request.urlopen(request)
    source = webpage.read()
    webpage.close()
    soup = bs.BeautifulSoup(source, 'html.parser')
    negritas = soup.find("body").find_all("b")
    listadoNegritas =[]    
    for val in negritas:
        listadoNegritas.append(val.text)
    
    print(f'-------------------------------------------------------')
    print(f'Página Web: {link}')
    print(f'-------------------------------------------------------')
    print(f'Palabras en negrilla con acentos españoles reemplazados por * :')
    print(f'-------------------------------------------------------')
        
    i=1    
    for palabra in listadoNegritas:
        procesado = re.sub(r"[áéíóú]", "*", palabra)
        print(f'{i}. {procesado}')
        i+=1
```
